```python
# Step 1: Import necessary libraries
from ultralytics import YOLO
import os
```

```python
# Step 2: Define dataset paths
dataset_dir = "C:/Users/User/Desktop/datasets/smart_nav_dataset"
train_images = os.path.join(dataset_dir, "images/train")
valid_images = os.path.join(dataset_dir, "images/val")
train_labels = os.path.join(dataset_dir, "labels/train")
valid_labels = os.path.join(dataset_dir, "labels/val")
```

```python
# Step 4: Load YOLOv8n model
model = YOLO("yolov8n.pt")  # You can switch to yolov8s.pt or yolov8m.pt
```

```python
import torch,gc
```

```python
torch.cuda.is_available()
```

```python
torch.cuda.empty_cache()
gc.collect()
```

```python
with open('smart_nav_dataset/data.yaml') as f:
    print(f.read())
```

```python
results = model.train(
    data='smart_nav_dataset/data.yaml',
    epochs=50,              
    batch=4,
    imgsz=640,
    device=0,
    workers=1,
    cache=False,
    amp=True,
    
    patience=20,
    save=True,
    save_period=10,
    
    project='runs/train',
    name='yolov8s_model',
    exist_ok=True,     
)
```

```python
# Step 7: Save/export trained model
model_path = os.path.join("runs/train/smart_nav_yolov8n", "best.pt")
model.export(format="onnx")  # Export example: ONNX
print(f"Model exported at: {model_path}")
```
